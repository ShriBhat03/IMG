```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
image=cv2.imread('naruto.jpg',0)
x,y=image.shape
z=np.zeros((x,y))
for i in range (0,x):
      for j in range (0,y):
            if(image[i][j]>50 and image[i][j]<150):
                z[i][j]=255
            else:
                z[i][j]=image[i][j]
equ=np.hstack((image,z))
plt.title('graylevel slicing with background')
plt.imshow(equ,'gray')
plt.show()
```
