```python
#image sharpen
from PIL import Image
from PIL import ImageFilter
import matplotlib.pyplot as plt
#load the image
my_image=Image.open('sasuke uchiha.jpg')
#use sharpen funn
sharp=my_image.filter(ImageFilter.SHARPEN)
#save the image
sharp.save('E:/image_sharpen.jpg')
sharp.show()
plt.imshow(sharp)
plt.show()
```

```python
#image flip
import matplotlib.pyplot as plt
#load the image
img=Image.open('sasuke uchiha.jpg')
plt.imshow(img)
plt.show()
#use the flip funn
flip=img.transpose(Image.FLIP_LEFT_RIGHT)

#save the image 
flip.save('E:/image_flip.jpg')
plt.imshow(flip)
plt.show()
```
